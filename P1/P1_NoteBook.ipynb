### ***Q.1***

```julia
function get_charge(worked_hours::T, part_cost::T) where {T<:Real}
    if worked_hours == 0 && part_cost != 0
        error("how it's possible??")
    elseif worked_hours == 0
        return 75
    else
        return max(120, (worked_hours * 75) + part_cost)
    end
end
```

```julia
function PrintCost()
    print("Enter \"Hours worked\"")
    HW = parse(Float64, readline())
    print("Enter \"Cost of parts\"")
    CP = parse(Float64, readline())
    Charge = get_charge(HW, CP)
    print("Charge = $Charge")
end
```

```julia
Hours_Worked = Int[0, 1, 2, 5, 10]
Cost_of_Parts = Int[0, 10, 20, 50, 100]
Charge = [get_charge(Hours_Worked[i], Cost_of_Parts[i]) for i ∈ 1:5]
DataFrame(
    "Hours Worked" => Hours_Worked,
    "Cost of Parts" => Cost_of_Parts,
    "Charge" => Charge
)
```

### ***Q.2***

```julia
function PrintSum()
    a = parse(Int, readprompt("a/b + c/d: Enter a"))
    b = parse(Int, readprompt("a/b + c/d: Enter b"))
    c = parse(Int, readprompt("a/b + c/d: Enter c"))
    d = parse(Int, readprompt("a/b + c/d: Enter d"))

    A = numerator(((a * d) + (c * b)) // (b * d))
    B = denominator(((a * d) + (c * b)) // (b * d))
    F = floor(Int, A / B)
    M = A % B
    display(L"%$F\frac{%$(Int(M))}{%$(Int(B))}")
end
```

```julia
# Input => a = 14, b = 4, c = 18, d = 10
PrintSum()
```

### ***Q.3***

```julia
function get_tax(income::T) where {T<:Real}
    restincome = 0.8 * income

    if restincome <= 15000
        tax = restincome / 0.1
    elseif restincome <= 35000
        tax = 1500 + (restincome - 15000) * 0.2
    else
        tax = 3500 + (restincome - 35000) * 0.25
    end
    return tax
end
```

```julia
function PrintTax()
    inc = parse(Float64, readprompt("Enter the income:"))
    tax = get_tax(inc)
    ATR = round(100 * tax / inc, digits = 2)
    print("Tax payable: $tax\$, Average tax rate: $ATR%")
end
```

```julia
# input = 27500
PrintTax()
```

```julia
# df = DataFrame(CSV.File(filename))
function generate_df()
    Ls = round.(Int, abs.(randn(400) .+ 5))
    Bs = round.(Int, abs.(randn(400) .+ 5))
    CpM = round.(Int, 10 * abs.(randn(400) .+ 4))
    df = DataFrame(
        "Room Number" => 1:400,
        "Length" => Ls,
        "Breadth" => Bs,
        "Cost" => CpM
    )
    return df
end
df = generate_df()
```

```julia
function get_newdf(df::DataFrame)
    return DataFrame(
        "Room Number" => 1:400,
        "Room Area" => df."Length" .* df."Breadth",
        "Cost" => df."Cost" .* (df."Length" .* df."Breadth")
    )
end
newdf = get_newdf(df)
```

```julia
mc, mrn = findmax(newdf."Cost")
println("The highest cost belonged to room $mrn and the cost was equal to $mc\$")
TotCost = sum(newdf."Cost")
println("Total cost: $TotCost\$")
```

```julia
Arr = rand(1:9, 5000)

function findSequence(Arr::Vector, value::Integer, index::Integer)
    if index == length(Arr) || Arr[index] != value
        return 0
    else
        return findSequence(Arr, value, index + 1) + 1
    end
end
maxnumber, repeatnumber = findmax(maximum(findSequence(Arr, i, j) for j ∈ 1:length(Arr)) for i ∈ 1:9)
print("The number: $maxnumber\nThe number of appearances: $repeatnumber")
```

```julia
function isPerfectSquare(n::Integer)
    for i ∈ 1:n
        if i * i == n
            return true
        end
    end
    return false
end
isPerfectSquare(36)
```

### ***Q.7***

```julia
function get_seq(n::Integer)
    Seq = Int[n]
    while n != 1
        if iseven(n)
            n = n ÷ 2
        else
            n = n * 3 + 1
        end
        push!(Seq, n)
    end
    return Seq
end
Seq = get_seq(13)
print("Length = $(length(Seq)),\nSequence = $Seq")
```

```julia
n = parse(Int, readprompt("Enter the beginning of the interval"))
m = parse(Int, readprompt("Enter the end of the interval"))

maxsizeseq, indx = findmax(length(get_seq(i)) for i ∈ n:m)
print("$(indx+n) generates the longest sequence of length $maxsizeseq")
```

### ***Q.8***

```julia
N = 10000
nums = (rand(2, N) .- 0.5) * 2
vecs = collect(zip(nums[1, :], nums[2, :]))
Norm = X -> √sum(x -> x .^ 2, X)
```

```julia
frac = length(findall(x -> x <= 1, Norm.(vecs))) / N * 4
δ = round(100 * abs(frac - π) / π, digits = 4)
print("Achieved π = $frac, δ = $δ%")
```

```julia
rectangle(w, h, x, y) = Shape(x .+ [0,w,w,0], y .+ [0,0,h,h])
P1 = begin
plot(sin.(0:0.0001:2π), cos.(0:0.0001:2π), c = :black)
plot!(rectangle(2, 2, -1, -1), opacity=0.2, fill = :steelblue)
end

P2 = begin
    scatter(vecs[findall(x -> x <= 1, Norm.(vecs))], ms=1, msw=0, c=:blue)
    scatter!(vecs[findall(x -> x > 1, Norm.(vecs))], ms=1, msw=0, c=:red)
end
plot(P1, P2, size=(1000, 500))
```

### ***Q.9***

```julia
function guess_the_number()
    span = [0, 100]

    for _ ∈ 1:3
        center = ((span[2] - span[1]) ÷ 2) + span[1]
        answer = readprompt("Is the number more than $center)? [y/n]")
        if answer == "y"
            span = [center, span[2]]
        elseif answer == "n"
            span = [span[1], center]
        else
            error("The input must be either y or n")
        end
    end
    randnum = rand(span[1]:span[2])
    answer = readprompt("Is the number equal to $randnum")
    if answer == "y"
        print("Successful")
    elseif answer == "n"
        print("Failed")
    else
        error("The input must be either y or n")
    end
end
guess_the_number()
```
